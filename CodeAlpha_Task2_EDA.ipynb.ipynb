# Key Findings

1. The dataset contains 9,551 restaurant records and 21 features.

2. New Delhi has the highest number of restaurants with 5,473 entries, followed by Gurgaon (1,118) and Noida (1,080).

3. The average restaurant rating is 2.67 out of 5.

4. The dataset contains information about restaurant locations, cuisines, ratings, customer votes, pricing and delivery services.

5. Customer votes vary significantly across restaurants, indicating differences in popularity.

6. Most restaurants are concentrated in a few major cities.

7. Restaurant ratings and customer votes can be used to identify highly preferred restaurants.

8. The dataset provides valuable insights into restaurant trends and customer preferences.

# Conclusion

Exploratory Data Analysis (EDA) was performed on the Zomato Restaurant Dataset using Python and Pandas.

The analysis showed that New Delhi dominates the dataset with the highest number of restaurants. The average restaurant rating was found to be 2.67, indicating moderate customer satisfaction across the dataset.

Various aspects such as restaurant popularity, pricing, cuisines, delivery services and customer ratings were explored. These findings help understand restaurant distribution and customer behavior patterns.

This analysis demonstrates how data can be cleaned, explored and interpreted to generate meaningful business insights.
